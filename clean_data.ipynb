## Clean the CSV data

```python
import pandas as pd
import os
data = pd.read_csv("/home/bimi/Bimi/users/Elahe/CARLA/data_collector/Collector/steering.csv")
display(data)
print(len(data))
```

```python
image_directory = '/home/bimi/Bimi/users/Elahe/CARLA/data_collector/Collector/_out/'
all_image_files = [file for file in os.listdir(image_directory) if file.endswith('.png')]  # Adjust the file extension as needed
print(len(all_image_files))
print(all_image_files)
```

```python
# Convert DataFrame to list
data_list = data["Image_Fname"].values.tolist()

# Display the resulting list
print(data_list)
```

```python
data_list1 = [d +'.png' for d in data_list]
print(data_list1)
```

```python
filtered_list1 = [item[0:-4] for item in data_list1 if item in all_image_files]
print(filtered_list1)
print(len(filtered_list1))
filtered_list2 = [image for image in all_image_files if image not in data_list1 ]
print(filtered_list2)
print(len(filtered_list2))
```

```python
# Drop rows where column 'Image_Fname' is not in the list 'values_to_keep'
df_filtered = data[data['Image_Fname'].isin(filtered_list1)]

# Display the filtered DataFrame
print("Filtered DataFrame:")
print(df_filtered)
display(df_filtered)

# Specify the path where you want to save the CSV file
csv_file_path = '/home/bimi/Bimi/users/Elahe/CARLA/data_collector/Collector/steering_cleaned.csv'

# Save the DataFrame to a CSV file
df_filtered.to_csv(csv_file_path, index=False)
```

## Cleaning the image data

```python
for image_to_delete in filtered_list2:
    image_path = os.path.join(image_directory, image_to_delete)
    os.remove(image_path)
    print(f"Deleted: {image_path}")
```
